```python
# Import von urlopen & BeautifulSoup
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# Zielseite definieren
import time
news_containersall =[]
for i in range(1,8):
    time.sleep (0)
    from urllib.request import urlopen as uReq
    from bs4 import BeautifulSoup as soup
    my_url = 'https://www.hnu.de/studium/news/' +str(i) 
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()
    soup = soup(page_html, "html.parser")
    
    
    news_containers =soup.findAll("li",{"class":"newslist__item"})
    
    news_containersall.append(news_containers)
    
```

```python
filename = "hnu_news.csv"
f = open(filename, "w")

headers = "date, header, content\n"

f.write(headers)

for i in range(0,7):
    for news_container in news_containersall[i]:
       
        content = news_container.p.text
        Header = news_container.h3.text
        
        # news_header = news_container.find("p",{"class":"news_title"}).text
        #news_long_ugly = news_container.find("p",{"class":"news_title"}).next_sibling.next_sibling
        #news_long_pretty = news_long_ugly[17:][:-4]
        print("Überschrift: " + Header)
        print("Text: " + content)
        
        #print("Text: " + news_long_pretty)

       # f.write(date + "," + news_header.replace(",", " ") + "," + news_long_pretty.replace(",", " ") + "\n")
    
f.close()
```
